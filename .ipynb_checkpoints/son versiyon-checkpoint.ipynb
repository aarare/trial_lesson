```python
import numpy as np
import glob
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
# Reading Multiple txt Files and Creating Data Frame

path = 'C:/Users/ACER/Desktop/dataset' # get path for all files that used by EDA C:/Users/ilke/Desktop/dataset

all_files = glob.glob(path + "/*.txt")

# create empty list and append all files

file_list = []

for filename in all_files:
    df = pd.read_csv(filename)
    file_list.append(df)

# Concatenate all data into one DataFrame that called "mta"
  
mta = pd.concat(file_list) 
```

```python
mta_top10_daily = mta.groupby(["station","dayname"]).sum()

mta_top10_daily = mta_top10_daily.reset_index()

mta_top10_daily = mta_top10_daily[(mta_top10_daily["station"]=="34 ST-PENN STA") 
                  | (mta_top10_daily["station"]=="23 ST")
                  | (mta_top10_daily["station"]=='FULTON ST')
                  | (mta_top10_daily["station"]=='GRD CNTRL-42 ST')
                  | (mta_top10_daily["station"]=="86 ST")
                  | (mta_top10_daily["station"]=="34 ST-HERALD SQ")
                  | (mta_top10_daily["station"]=="42 ST-PORT AUTH")
                  | (mta_top10_daily["station"]=="125 ST")             
                  | (mta_top10_daily["station"]=="59 ST")
                  | (mta_top10_daily["station"]=="TIMES SQ-42 ST")]            

mta_top10_daily = mta_top10_daily.reset_index()

mta_top10_daily
```

```python
#rearranging the indices 
mta_top10_daily_order = mta_top10_daily.reindex(index = [1,5,6,4,0,2,3,8,12,13,11,7,9,10,15,19,20,18,14,16,17,22,26,27,25,21,23,24,29,33,34,32,28,30,31,36,40,41,39,35,37,38,43,47,48,46,42,44,45,50,47,48,46,49,51,52,57,54,55,53,56,58,59,64,61,62,60,63,65,66,68,69,67])

mta_top10_daily_order.head(10)
```

```python
#pivots for the heatmap

mta_top10_heatmap = mta_top10_daily_order.groupby(["station","dayname"]).traffic.mean().reset_index()

mta_top10_heatmap = mta_top10_heatmap.pivot("dayname","station","traffic")

mta_top10_heatmap = mta_top10_heatmap.reindex(index = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])

mta_top10_heatmap
```

```python
#plotting the heatmap
plt.figure(figsize=(15,11),dpi=200),
sns.set(font_scale=1.5)
sns.heatmap(mta_top10_heatmap, linewidths=2, cmap="Greens",cbar_kws={'label': 'Traffic'})
plt.style.use('default')
plt.ylabel("Days")
plt.xlabel("Stations")
plt.yticks(weight="bold",rotation=0, size=12)
plt.xticks(weight="bold",rotation=25,size=10)

plt.title("Top 10 Busiest Stations / Day",weight="bold",c="Red", fontsize=15)


plt.show()
```

# Busiest days and hours with respect to a station as a heatmap

```python
mta_time = mta.sort_values(["station","turnstile","datetime"]).reset_index()
mta_time_group = mta_time.groupby(["turnstile"])
mta_time["entries_count"] = mta_time_group["entries"].transform(pd.Series.diff)
mta_time["counts_exits"] = mta_time_group["exits"].transform(pd.Series.diff)
mta_time.head(10)
```

```python
print("Descriptive Statistics for entries_count column:")
print(mta_time["entries_count"].describe())
print("Descriptive Statistics for counts_exits column:")
print(mta_time["counts_exits"].describe())
```

```python
print("Negative values for entries_count column: ")
print(len(mta_time[mta_time["entries_count"]<0]))
print("Negative values for counts_exits column:")
print(len(mta_time[mta_time["counts_exits"]<0]))
```

```python
mta_time["entries_count"] = mta_time["entries_count"].fillna(0)
mta_time["counts_exits"] = mta_time["counts_exits"].fillna(0)
mta_time['entries_count'][mta_time['entries_count'] < 0] = (mta_time["entries_count"].median())
mta_time['counts_exits'][mta_time['counts_exits'] < 0] = (mta_time["counts_exits"].median())
mta_time['entries_count'][mta_time['entries_count'] > (mta_time["entries_count"].quantile(0.95))] = (mta_time["entries_count"].median()) 
mta_time['counts_exits'][mta_time['counts_exits'] > (mta_time["counts_exits"].quantile(0.95))] = (mta_time["counts_exits"].median())
mta_time
```

```python
print("Descriptive Statistics for entries_count column:")
print(mta_time["entries_count"].describe())
print("Descriptive Statistics for counts_exits column:")
print(mta_time["counts_exits"].describe())
```

```python
mta_time["traffic"] = mta_time["entries_count"] + mta_time["counts_exits"]
mta_time.head(10)
```

```python
mta_time.groupby(["station"]).sum().sort_values("traffic",ascending = False).head(10)
```

```python
mta_time_a=mta_time[mta_time["station"]=="34 ST-PENN STA"]
stations_day_time = mta_time_a[['station', 'time', 'dayname','traffic']]
stations_day_time_group = stations_day_time.groupby(['station','dayname','time'], as_index=False)
stations_day_time_group = stations_day_time_group[['traffic']].sum()
stations_day_time_group['time'] = pd.to_datetime(stations_day_time_group['time'], format='%H:%M:%S')
rounded_day_time = stations_day_time_group.set_index('time').groupby(['station','dayname'])
rounded_day_time = rounded_day_time.resample("4H").sum()
rounded_day_time = rounded_day_time.reset_index()
rounded_day_time.head(10)
```

```python
mta_top1_heatmap_detail = rounded_day_time.groupby(["dayname","time"])
mta_top1_heatmap_detail = mta_top1_heatmap_detail["traffic"].mean().reset_index()

mta_top1_heatmap_detail = rounded_day_time.pivot(index="time", columns="dayname",values="traffic")
mta_top1_heatmap_detail = mta_top1_heatmap_detail[["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]]
```

```python
#plotting the heatmap
plt.figure(figsize=(15,8),dpi=200),
sns.set(font_scale=1.5)
sns.heatmap(mta_top1_heatmap_detail,annot=False, linewidths=2, fmt = "", cmap="Reds",cbar_kws={'label': 'Traffic'})
plt.style.use('seaborn-talk')
plt.ylabel("Hours")
plt.xlabel("Days")
plt.yticks(weight="bold",rotation=0)
plt.yticks([0,1,2,3,4,5,6],["00:00","04:00","08:00","12:00","16:00","20:00","00:00"])
plt.xticks(weight="bold",rotation=0)
plt.title("Busiest Day and Time for 34 ST-PENN Station",weight="bold",c="red", fontsize=15)
ax.xaxis.get_majorticklabels()[2].set_x(-.5)
plt.show()
```
